```python
import pandas
import numpy as np
import matplotlib.pyplot as plt



path = '/Users/vietld/PycharmProjects/har/data/001/in/data.csv'
with open(path, 'r') as f:
    lines = f.readlines()
data=[]
for line in lines:
    parts = line.strip().split(',')
    _d = (int(parts[0]),[float(x) for x in parts[1:-1]], parts[-1])
    data.append(_d)

print('Number of samples: ')
len(data)
```

```python
labels = set([x[2]for x in data])
print(labels)
```

```python
def plot_data(data, label, start=0, end=None,dim=0):
    #print(label)
    data = [x for x in data if x[2]==label]
    if end:
        data = data[start:end]
    else:
        data = data[start:]
    tp = [x[0] for x in data ]
    start = tp[0]
    tp = [(x-start)/20 for x in tp]

    val = [x[1][dim] for x in data]    
    plt.plot(tp, val)
    plt.show()
```

```python
plot_data(data, 'brushing', start=0, end=200, dim=0)
```
